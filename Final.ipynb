We looked at neighborhood-level arrest data published in the Community Profiles and sourced from the City of Pittsburgh Police Department's Annual Report of Major Crimes. "Major crimes" include murder, rape, robbery, assault, burglary, and larceny. Generally, a neighborhood is considered more safe with a lower crime rate.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl

data = pd.read_csv("Arrests for Major Crimes.csv")
average_crimerate = data.groupby('neighborhood')['overall_crime_rate'].mean().reset_index()
average_crimerate = average_crimerate.set_index('overall_crime_rate').sort_index(ascending=True).reset_index()
#Rename the first (index) column to "rank" and start rank from 1 instead of 0
average_crimerate.index = average_crimerate.index + 1
average_crimerate = average_crimerate.rename_axis("rank")
highest_crime_neighborhood = average_crimerate.iloc[-1]
nameMax = highest_crime_neighborhood['neighborhood']
crimeMax = highest_crime_neighborhood['overall_crime_rate']
print("The neighborhood with the most crime is " + nameMax + " with an overall crime rate of " + crimeMax)
count.head()
```
