# Xavier Hilmey: Major Crimes

We looked at neighborhood-level arrest data published in the Community Profiles and sourced from the City of Pittsburgh Police Department's Annual Report of Major Crimes. "Major crimes" include murder, rape, robbery, assault, burglary, and larceny. Generally, a neighborhood is considered more safe with a lower crime rate.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl

data = pd.read_csv("Arrests for Major Crimes.csv")
average_crimerate = data.groupby('neighborhood')['overall_crime_rate'].mean().reset_index()
average_crimerate = average_crimerate.set_index('overall_crime_rate').sort_index(ascending=True).reset_index()
count = average_crimerate['neighborhood'].value_counts()
#Rename the first (index) column to "rank" and start rank from 1 instead of 0
average_crimerate.index = average_crimerate.index + 1
average_crimerate = average_crimerate.rename_axis("rank")
nameMax = count.max()
crimeMax = count.idxmax()
print(f"The neighborhood with the most crime is{nameMax} with an overall crime rate of {crimeMax}")
print(average_crimerate)
```

It seems as if there is a direct correlation between the general economic status of a neighborhood and the crime rate of said neighborhood. 

# Graph

```python
average_crimerate = average_crimerate.set_index('overall_crime_rate').sort_index(ascending=False).reset_index()
pl.figure(figsize=(14, 8))
pl.bar(average_crimerate['neighborhood'], average_crimerate['overall_crime_rate'], color = 'blue')
pl.ylabel('Average Overall Crime Rate')
pl.xlabel('Neighborhood')
pl.title('Average Overall Crime Rate per Neighborhood', fontsize = 14)
pl.xticks(rotation = 90)
pl.show()
```

From what the graph shows us, the crime is very concentrated in certain areas, while the rest of the neighborhoods are fairly similar in crime rate.
